#### 1. Bar Chart: Population by State

```python
df = df.sort_values(by="Population", ascending=False)  # Ascending order for correct bar orientation
plt.figure(figsize=(12, 8))
sns.barplot(y="State", x="Population", data=df, palette="viridis")
plt.xlabel("Population")
plt.ylabel("State")
plt.title("Population Distribution by State")
plt.show()
```

#### 2. Bar Chart: Gender Distribution by State

```python
plt.figure(figsize=(15, 8))
plt.bar(df["State"],df["Males"],color="red", label="Males")
plt.bar(df["State"],df["Females"], bottom = df["Males"], label="Females",color = "green")
plt.xlabel("State")
plt.ylabel("Population")
plt.title("Gender Distribution by State")
plt.xticks(rotation=90)
plt.legend(["Males", "Females"])
plt.show()
```

#### 3. Histogram: Population Density Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Density (1/km*km)"], kde=True, color="Green")
plt.xlabel("Population Density (per km²)")
plt.ylabel("Frequency")
plt.title("Distribution of Population Density")
plt.show()
```
